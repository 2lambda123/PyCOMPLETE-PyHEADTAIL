```python
import numpy as np
from scipy.constants import m_p, c, e
```

```python
from PyHEADTAIL.particles.particles import Particles
from PyHEADTAIL.particles.generators import Gaussian6DTwiss
from PyHEADTAIL.trackers.simple_long_tracking import RFSystems
import PyHEADTAIL.aperture.aperture as aperture
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# machine parameters
circumference = 157.
inj_alpha_x = 0
inj_alpha_y = 0
inj_beta_x = 5.9 # in [m]
inj_beta_y = 5.7 # in [m]
Qx = 5.1
Qy = 6.1
gamma_tr = 4.05
alpha_c_array = [gamma_tr**-2]
V_rf = 8e3 # in [V]
harmonic = 1
phi_offset = 0 # measured from aligned focusing phase (0 or pi)
pipe_radius = 5e-2

# beam parameters
Ekin = 1.4e9 # in [eV]
intensity = 1.684e12
epsn_x = 2.5e-6 # in [m*rad]
epsn_y = 2.5e-6 # in [m*rad]
epsn_z = 1.2 # 4pi*sig_z*sig_dp (*p0/e) in [eVs]

# calculations
gamma = 1 + e * Ekin / (m_p * c**2)
beta = np.sqrt(1 - gamma**-2)
eta = alpha_c_array[0] - gamma**-2
if eta < 0:
    phi_offset = np.pi - phi_offset
Etot = gamma * m_p * c**2 / e
p0 = np.sqrt(gamma**2 - 1) * m_p * c
Qs = np.sqrt(np.abs(eta) * V_rf / (2 * np.pi * beta**2 * Etot))
beta_z = np.abs(eta) * circumference / (2 * np.pi * Qs)
```

```python
def plot_phase_space(bunch, ax0, ax1, ax2, col):
    # phase spaces
    ax0.scatter(bunch.x, bunch.xp, color=col)
    ax1.scatter(bunch.y, bunch.yp, color=col)
    ax2.scatter(bunch.z, bunch.dp, color=col)
    # statistical quantities
    ax0.scatter(bunch.mean_x(), bunch.mean_xp(), color='red')
    ax1.scatter(bunch.mean_y(), bunch.mean_yp(), color='red')
    ax2.scatter(bunch.mean_z(), bunch.mean_dp(), color='red')
```

```python
def generate_bunch(n_particles):
    bunch = Gaussian6DTwiss(
        n_particles, intensity, e, m_p, circumference, gamma,
        inj_alpha_x, inj_beta_x, epsn_x,
        inj_alpha_y, inj_beta_y, epsn_y,
        beta_z, epsn_z
        ).generate()
    return bunch
```

```python
def check_macroparticlenumber(bunch):   
    if bunch.macroparticlenumber == np.sum(bunch.alive_all):
        print 'bunch.macroparticlenumber was updated correctly.'
        print 'bunch.macroparticlenumber new:', bunch.macroparticlenumber
        print 'fraction lost [%]:', (1. - float(bunch.macroparticlenumber)/float(bunch.macroparticlenumber_all)) * 100.
    else:
        print 'bunch.macroparticlenumber was not updated. Check state of apply_losses_here for the specified apertures.'
```

```python
# (I) RectangularApertureX
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

apt_x = aperture.RectangularApertureX(x_low=-0.004, x_high=0.005, apply_losses_here=True)
apt_x.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```

```python
# (II) RectangularApertureY
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

apt_y = aperture.RectangularApertureY(y_low=-0.005, y_high=0.005, apply_losses_here=True)
apt_y.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```

```python
# (III) RectangularApertureZ
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

apt_z = aperture.RectangularApertureZ(z_low=-15, z_high=25, apply_losses_here=True)
apt_z.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```

```python
# (IV) CircularApertureXY
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

apt_xy = aperture.CircularApertureXY(radius=0.003, apply_losses_here=True)
apt_xy.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```

```python
# (V) RFBucketAperture
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

rfsystems = RFSystems(circumference, [harmonic], [V_rf], [phi_offset],
                      alpha_c_array, gamma)
apt_rf = aperture.RFBucketAperture(is_accepted=rfsystems.rfbucket.make_is_accepted(margin=0.9),
                                   apply_losses_here=True)
apt_rf.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```

```python
# (VI) RectangularApertureX, without applying losses
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

apt_x = aperture.RectangularApertureX(x_low=-0.004, x_high=0.005, apply_losses_here=False)
apt_x.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```

```python
# (VII) RectangularApertureX and RectangularApertureZ. Apply losses once only.
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(7, 7))

bunch = generate_bunch(1000)
plot_phase_space(bunch, ax0, ax1, ax2, col='b')

apt_x = aperture.RectangularApertureX(x_low=-0.004, x_high=0.005, apply_losses_here=False)
apt_z = aperture.RectangularApertureZ(z_low=-10, z_high=20, apply_losses_here=True)
apt_x.track(bunch)
apt_z.track(bunch)
check_macroparticlenumber(bunch)

plot_phase_space(bunch, ax0, ax1, ax2, col='g')
plt.show()
```
